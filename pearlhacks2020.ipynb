```python
import pandas as pd
import numpy as np

our_file = project.get_file("StormEvents_details-ftp_v1.0_d2018_c20200219.csv") # returns BytesIO object from file name
our_df = pd.read_csv(our_file) # parse to a Pandas DataFrame using pd.read_csv()
print(type(our_df))
our_df.head()
```

```python
newvalues = []
for row in our_df['OUR_METRIC']:
    if row <= 0.001:
        newvalues.append(1)
    elif row <= 1:
        newvalues.append(2)
    elif row <= 5:
        newvalues.append(3) 
    elif row <= 10:
        newvalues.append(4)
    elif row <= 20:
        newvalues.append(5)
    elif row <= 30:
        newvalues.append(6)
    elif row <= 40:
        newvalues.append(7)
    elif row <= 60:
        newvalues.append(8)
    elif row <= 80:
        newvalues.append(9)
    else:
        newvalues.append(10)
print(newvalues)
```

```python
np_array = np.asarray(newvalues)
our_df['Severity'] = np_array
```
